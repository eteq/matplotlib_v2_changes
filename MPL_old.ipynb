```python
%matplotlib inline
import matplotlib
from matplotlib import pyplot as plt
matplotlib.__version__
```

```python
np.random.seed(12345) # that's amazing I have the same combination on my luggage
```

```python
plt.figure(figsize=(12, 6))

x = np.linspace(-2,2,1000)
y1 = np.random.randn(len(x))+np.exp(-(x+0.25)**2/2)*10
y2 = np.random.randn(len(x))+np.exp(-(x-0.5)**2/2/(0.5))*10
y3 = np.random.randn(len(x))+np.exp(-(x-0.5)**2/2/(0.1))*10
plt.plot(x, y1, label='a')
plt.plot(x, y2, label='b')
plt.plot(x, y3, label='c')

plt.legend()
```

```python
x = np.random.rand(1000)*4-1
y = np.random.rand(1000)*4

#rosenbrock banana
z = np.log((1-x)**2 + 100*(y-x)**2)

plt.scatter(x, y, c=z, label='d')
plt.colorbar().set_label('rosenbrock banana', fontsize=18)

plt.xlabel('$\\sqrt{x+\\alpha}$', fontsize=18)
plt.ylabel('y $y$', fontsize=18)

plt.legend()
```

```python
fig, axs = plt.subplots(2, 2)

x = np.linspace(-2,2,1000)
y = np.random.randn(len(x))+np.exp(-(x+0.25)**2/2)*10
for ax in axs.flat:
    ax.plot(x, y)
plt.tight_layout()
```

```python
fig, axs = plt.subplots(2, 2, sharex=True)

x = np.linspace(-2,2,1000)
y = np.random.randn(len(x))+np.exp(-(x+0.25)**2/2)*10
for ax in axs.flat:
    ax.plot(x, y)
plt.subplots_adjust(hspace=0)
```
